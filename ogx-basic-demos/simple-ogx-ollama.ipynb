# Runing a simple LLM Model using OGX and OLLAMA

Global Function to Use

```python
import requests

OGX = "http://localhost:8321/v1"

def call_model(model, prompt):
    r = requests.post(f"{OGX}/responses", json={
        "model": model,
        "input": prompt
    })
    return r.json()["output"][0]["content"][0]["text"]
```

## Use llama3.2:3b Model

```python
print("LLAMA RESULT:\n")

print(call_model(
    "ollama/llama3.2:3b",
    "Explain Kubernetes in one short paragraph"
))
```

## Use deepseek-r1:8b Model

```python
print("\nDEEPSEEK RESULT:\n")

print(call_model(
    "ollama/deepseek-r1:8b",
    "Explain Kubernetes in one short paragraph"
))
```
